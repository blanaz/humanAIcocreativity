```python
def plot_distance(distances, i):
    # Plot the evolution of distances for each session
    plt.figure(figsize=(20, 10))
    distances_session = distances.distance.iloc[i]
    submissions_session = distances.submissions.iloc[i]
    x = range(len(distances_session))
    plt.plot(x, distances_session)

    # add a red vertical line for each submission
    for j in range(len(submissions_session)):
        if submissions_session[j] == 1:
            plt.axvline(x=j, color='r')
```

```python
def plot_distance(distances, i):

    # Plot the evolution of distances for each session
    plt.figure(figsize=(30, 10))
    distances_session = distances.distance.iloc[i]
        # Calculate the moving average
    moving_average = pd.Series(distances_session).rolling(3).mean()
    submissions_session = distances.submissions.iloc[i]
    x = range(len(distances_session))
    plt.plot(x, distances_session, alpha = 0.2)

    # add a red vertical line for each submission
    for j in range(len(submissions_session)):
        if submissions_session[j] == 1:
            plt.axvline(x=j, color='r')

    # Plot the moving average on top of the distances
    plt.plot(x, moving_average)
    plt.title("Participant " + str(i + 1),fontsize = 38)
    plt.xlabel("Time", fontsize = 30)
    plt.ylabel("Total slider adjustment since last image generation", fontsize = 30)
    plt.tick_params(labelsize=25)
   
```

```python
for i in range(6):
    plot_distance(distances_new, i)
```
